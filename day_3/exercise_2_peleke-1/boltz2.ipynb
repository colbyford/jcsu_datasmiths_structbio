```python
output_path = ''
```

```python
## Define function to create Boltz-2 YAML configuration for multimer folding
def create_boltz_yaml(antibody_sequence: str, antigen_sequence: str, output_path: str):
    config_json = {
        "version": 1,
        "sequences": [
            {
            "protein": {
                "id": "B",
                "sequence": antibody_sequence.replace('|', 'GGGSGGGSGGGS') ## Make scFV sequence
            }
            },
            {
            "protein": {
                "id": "A",
                "sequence": antigen_sequence.replace('[', '').replace(']', '') ## Remove brackets
            }
            }
        ]
        }
    
    ## Make directories if they don't exist
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    
    ## Write YAML file
    with open(output_path, 'w') as f:
        yaml.dump(config_json, f)
```

```python
## Define a function for running Boltz-2 folding from the command line
def run_boltz_folding(yaml_config_path: str, output_dir: str, accelerator: str = 'gpu'):
    boltz_command = f"boltz predict {yaml_config_path} --out_dir {os.path.dirname(output_dir)} --write_full_pae --accelerator {accelerator}"
    os.system(boltz_command)
```

```python
yaml_config_path = f"{output_path}/antibody_antigen_001.yaml"

## Create YAML file
create_boltz_yaml(antibody_sequence, antigen_sequence, yaml_config_path)

## Run Boltz folding
run_boltz_folding(yaml_config_path, output_path, accelerator='gpu')
```
